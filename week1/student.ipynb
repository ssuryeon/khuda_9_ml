```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset repository path
url = "https://raw.githubusercontent.com/gaeng02/KHUDA-9th-ML/main/week1/titanic.csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
### answer init
answer = [0] * 6
```

## Task 1
df의 feature중 `Fare`의 중앙값을 구하고, 해당 값을 `answer[1]`에 저장하세요.

```python
## Input Box

median_fare = 0

median_fare = df['Fare'].median()


print(median_fare)
```

```python
answer[1] = median_fare
```

## Task 2
df의 feature중 `Pclass`에 대해 평균 생존률이 가장 높은 `Pclass`를 `answer[2]`에 저장하세요.

```python
## Input Box

high_mean_survival_rate = 0

df_groupby = df.groupby(by='Pclass')[['Survived']].mean()
max_row = df_groupby[df_groupby['Survived'] == df_groupby['Survived'].max()]

high_mean_survival_rate = 1
print(max_row)
```

```python
answer[2] = high_mean_survival_rate
```

## Task 3
df의 feature중 `Age`에 대해서 `Age < 10`인 승객 수를 `answer[3]`에 저장하세요.

```python
## Input Box

age_under_10 = 0
filtered_df = df[df['Age'] < 10]

age_under_10 = filtered_df['Name'].count()
print(age_under_10)
```

```python
answer[3] = age_under_10
```

## Task 4
성별이 여성(`Sex`가 `female`)이면서 생존자(`Survived`가 `1`)인 승객 수를 `answer[4]`에 저장하세요.

```python
## Input Box

female_survive_rate = 0

filtered_df = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
female_survive_rate = filtered_df['Name'].count()

print(female_survive_rate)
```

```python
answer[4] = female_survive_rate
```

## Task 5
다음 그래프가 df의 feature중 어떤 feature의 그래프인지 시각화를 해본 후, feature 명을 `answer[5]`에 저장하세요.

다운로드.png

```python
## Input Box

feature_name = "Fare"

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df[feature_name], bins=10)

print(feature_name)
plt.show()
```

```python
answer[5] = feature_name
```

```python
### 정답 채점 코드
import hashlib

ANSWER_URL = "https://raw.githubusercontent.com/gaeng02/KHUDA-9th-ML/main/week1/answer.csv"

def md5 (s) : return hashlib.md5(s.encode("utf-8")).hexdigest()

ans_df = pd.read_csv(ANSWER_URL)
gt = {int(r["task"]): str(r["answer"]).strip().lower()
      for _, r in ans_df.iterrows()}

wrong = []

for i in range(1, 6) :
    user_answer = "" if answer[i] is None else md5(str(answer[i]).strip().lower())
    if (user_answer != gt.get(i, "")) : wrong.append(i)

print("탈출!" if not wrong else f"틀린번호 : {wrong}")
```
